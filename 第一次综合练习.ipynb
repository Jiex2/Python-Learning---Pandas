## 1. 企业收入的多样性

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/company.csv')
df2=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/company_data.csv')
df1.head(3)
```

```python
df2.head(3)
```

```python
#修改证券代码和日期的格式
df2['证券代码']=df2['证券代码'].map(lambda x: '#'+np.str(x).zfill(6))
df2['日期']=pd.to_datetime(df2['日期'])
df2['日期']=df2['日期'].dt.year
```

```python
df2.head(3)
```

```python
#按照不同公司（证券代码）和年份汇总收入
df2_sum=pd.DataFrame(df2.groupby(['证券代码','日期'])['收入额'].sum())
df2_sum=df2_sum.reset_index().rename(columns={'收入额':'年总收入额'})
df2_sum.head(3)
```

```python
#将汇总收入与df2原始表对应
df2_merge=df2.merge(df2_sum, on=['证券代码','日期'], how='left')
df2_merge.head(3)
```

```python
#计算收入指数
p=df2_merge['收入额']/df2_merge['年总收入额']
df2_merge['收入指数']=-(p*p.apply(np.log))
df2_sumindex=pd.DataFrame(df2_merge.groupby(['证券代码','日期'])['收入指数'].sum().reset_index())
df2_sumindex.head(3)
```

```python
#将计算好的指数表与df1匹配
df1.merge(df2_sumindex, on=['证券代码','日期'], how='left')
```
