```python
df1=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/company.csv')
df2=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/company_data.csv')
df1.head(3)
```

```python
#ä¿®æ”¹è¯åˆ¸ä»£ç å’Œæ—¥æœŸçš„æ ¼å¼
df2['è¯åˆ¸ä»£ç ']=df2['è¯åˆ¸ä»£ç '].map(lambda x: '#'+np.str(x).zfill(6))
df2['æ—¥æœŸ']=pd.to_datetime(df2['æ—¥æœŸ'])
df2['æ—¥æœŸ']=df2['æ—¥æœŸ'].dt.year
```
