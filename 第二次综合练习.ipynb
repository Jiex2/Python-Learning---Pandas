## 任务4: 显卡日志

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_table('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/benchmark.txt', skiprows=12, nrows=382, header=None, sep=' ', engine='python')
df=df.reset_index()
df.head()
```

```python
df_concat=pd.concat([pd.DataFrame(df.values[::2]), pd.DataFrame(df.values[1::2])], axis=1)
df_concat.columns=range(0,df_concat.shape[1])
df_concat.head()
```

```python
df_final=df_concat[[1,2,5,18]]
df_final.columns=['Status', 'Precision', 'Model', 'Time']
df_final.head()
```

```python
df_final=df_final.sort_index()
df_final.loc[:,'Time']=round(df_final['Time'].astype('float'), 3)
df_final.head()
```

```python
df_final1=df_final.pivot_table(index='Model', columns=['Status', 'Precision'], values='Time', aggfunc='mean')
df_final1.head()
```

```python
df_final1.columns=df_final1.columns.map(lambda x: x[0]+' '+x[1])
df_final1=df_final1.reset_index()
df_final1.head()
```

```python
df_final1.sort_values('Model')
```

## 任务5: 水压站点的特征工程

```python
df1=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/yali18.csv')
df2=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/yali19.csv')
df3=pd.read_csv('/Users/jie/Documents/Python/joyful-pandas-master/data/ex/qx1819.csv')
```

```python
#1.
df12=df1.append(df2)
df12_melt=df12.melt(id_vars=['Time','MeasName'], value_vars=df12.columns[2:], var_name='Hour', value_name='P')
df12_melt.head()
```

```python
df12_melt['Hour']=df12_melt['Hour'].str.replace('H', ' ', regex=True)
df12_melt.head()
```

```python
a=df12_melt['Time']+' '+df12_melt['Hour'].str.strip().str.zfill(2)
df12_melt['Time']=pd.to_datetime(a, format='%Y-%m-%d %H')
df12_melt
```

```python
df12_melt['MeasName']=df12_melt['MeasName'].str.replace('站点', ' ', regex=True).str.strip().astype('int64')
df12_final=df12_melt.drop(['Hour'], axis=1).rename(columns={'MeasName':'站点', 'P':'压力'}).set_index('Time').sort_values(['Time','站点'])
df12_final
```

```python
#2.
#2.1
#两种C的表示都只占一个字符
df3_1=df3.copy()
df3_1[['最高温','最低温']]=df3_1['气温'].str.split('～', expand=True)
df3_1['最高温']=df3_1['最高温'].str.strip()
df3_1['最低温']=df3_1['最低温'].str.strip()
df3_1['温差']=pd.to_numeric([df3_1.loc[i,'最高温'][:-1] for i in range(0,df3_1.shape[0])])-pd.to_numeric([df3_1.loc[i,'最低温'][:-1] for i in range(0,df3_1.shape[0])])
df3_1.head()
```

```python
#2.2
df3_2=df3_1.copy()
df3_2['沙暴']=df3_2['天气'].str.contains('沙暴')
df3_2['雾']=df3_2['天气'].str.contains('雾')
df3_2['雨']=df3_2['天气'].str.contains('雨')
df3_2['雪']=df3_2['天气'].str.contains('雪')
df3_2['晴']=df3_2['天气'].str.contains('晴')
df3_2.head()
```

```python
#2.3
df3_3=df3_2.copy()
gb=df3_3.groupby(['天气'])
gb.groups.keys()
```

```python
#选择最后一个天气状态评估雨量或雪量
b=df3_3['天气'].str.split('转', expand=True)
b[1]=b[1].fillna(b[0])
df3_3['雨量']=b[1].str.extract('(\w*雨)')
df3_3['雪量']=b[1].str.extract('(\w*雪)')
df3_3
```

```python
#2.4
df3_4=df3_3.copy()
df3_4['东风']=df3_4['风向'].str.contains('东').astype('int64')
df3_4['西风']=df3_4['风向'].str.contains('西').astype('int64')
df3_4['南风']=df3_4['风向'].str.contains('南').astype('int64')
df3_4['北风']=df3_4['风向'].str.contains('北').astype('int64')
df3_4.head()
```
